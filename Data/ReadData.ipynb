```python
import numpy as np

def split_algorithm(giant_tour, demands, Q, cost_matrix):
    """
    Implements the Split algorithm to split a giant tour into feasible routes.
    
    Parameters:
    - giant_tour: List of customer indices in the giant tour.
    - demands: List of demands for each customer (indexed like giant_tour).
    - Q: Vehicle capacity.
    - cost_matrix: 2D array where cost_matrix[i][j] gives the cost of traveling from i to j.
    
    Returns:
    - V: Array of minimum costs to reach each customer.
    - P: Array of predecessors for reconstructing the solution.
    """
    n = len(giant_tour)  # Number of customers in the giant tour
    V = [float('inf')] * (n + 1)  # Cost to reach each customer (DP table)
    P = [-1] * (n + 1)  # Predecessor array for backtracking
    V[0] = 0  # Cost to start at depot
    
    for i in range(1, n + 1):
        j = i
        load = 0
        cost = 0
        
        while j <= n and load + demands[giant_tour[j - 1] - 1] <= Q:  # Adjust index
            load += demands[giant_tour[j - 1] - 1]  # Adjust index
            if i == j:
                cost = cost_matrix[0][giant_tour[j - 1]] + cost_matrix[giant_tour[j - 1]][0]  # Cost from depot
            else:
                cost += (cost_matrix[giant_tour[j - 2]][giant_tour[j - 1]]
                         + cost_matrix[giant_tour[j - 1]][0]
                         - cost_matrix[giant_tour[j - 2]][0])
                
            # Update DP table if better solution found
            if V[i - 1] + cost < V[j]:
                V[j] = V[i - 1] + cost
                P[j] = i - 1  # Store the predecessor
                
            j += 1  # Move to the next customer
            
    return V, P


def extract_solution(giant_tour, P):
    """
    Extracts the routes from the predecessor array after the Split algorithm.
    
    Parameters:
    - giant_tour: List of customer indices in the giant tour.
    - P: Predecessor array from the Split algorithm.
    
    Returns:
    - routes: List of routes (each route is a list of customers).
    """
    routes = []  # List to store routes
    j = len(P) - 1  # Start from the last customer
    
    while j > 0:
        route = []  # Current route
        for k in range(P[j] + 1, j + 1):
            route.append(giant_tour[k - 1])
        routes.insert(0, route)  # Add route at the beginning
        j = P[j]  # Move to the predecessor
    
    return routes


# Example Usage
if __name__ == "__main__":
    # Example inputs
    giant_tour = [1, 2, 3, 4, 5]  # Customer indices
    demands = [5, 6, 7, 4, 8]     # Demands for customers
    Q = 15                        # Vehicle capacity
    cost_matrix = [               # Cost matrix (depot is index 0)
        [0, 10, 20, 30, 40, 50],
        [10, 0, 12, 15, 25, 35],
        [20, 12, 0, 18, 22, 28],
        [30, 15, 18, 0, 14, 16],
        [40, 25, 22, 14, 0, 12],
        [50, 35, 28, 16, 12, 0]
    ]

    # Run Split algorithm
    V, P = split_algorithm(giant_tour, demands, Q, cost_matrix)

    # Extract routes
    routes = extract_solution(giant_tour, P)
    
    # Print results
    print("Optimal Cost to Reach Each Customer:", V[1:])
    print("Predecessors for Backtracking:", P[1:])
    print("Extracted Routes:")
    for idx, route in enumerate(routes):
        print(f"Route {idx + 1}: {route}")
```
